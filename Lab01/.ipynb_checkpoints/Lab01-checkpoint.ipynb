```python
# Pandas for managing datasets
import pandas as pd
```

```python
# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
# Seaborn for plotting and styling
import seaborn as sns
```

```python
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
df.head()
```

```python
# Select relevant columns
selected_columns = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
```

```python
# Calculate correlations
correlation_matrix = df[selected_columns].corr()
```

```python
pearson_corr = correlation_matrix.corr(method='pearson')
spearman_corr = correlation_matrix.corr(method='spearman')
```

```python
# Pearson correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', center=0)
plt.title('Pearson Correlation Coefficient')
plt.show()
```

```python
# Spearman correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', center=0)
plt.title('Spearman Rank Correlation')
plt.show()
```

# Task 2

```python
dataset2 = pd.read_csv('Au_nanoParticle_dataset.csv')
```

```python
print(dataset2.head(20))
```

```python
# Filter columns [N_total, N_bulk, N_surface, R_avg]
filtered_dataset2 = dataset2[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python
# Display the first 20 samples of the new DataFrame
print(filtered_dataset2.head(20))
```

```python
# Calculate mean, standard deviation, and quartile values for each column
statistics = filtered_dataset2.describe()
print(statistics)
```

```python
# Plot histograms of the features in a 1x4 layout
plt.figure(figsize=(15, 4))

for i, column in enumerate(filtered_dataset2.columns):
    plt.subplot(1, 4, i + 1)
    plt.hist(filtered_dataset2[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(column)
    plt.xlabel('Value')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Visualize scatter plots and histograms
sns.pairplot(filtered_dataset2)
plt.show()
```

```python
# Create a PairGrid with modified diagonal and lower plots
g = sns.PairGrid(filtered_dataset2)

# Bivariate histogram

g.map_upper(sns.histplot)

 # Histogram with KDE on diagonal
    
g.map_diag(sns.histplot, kde=True) 

 # Bivariate KDE
    
g.map_lower(sns.kdeplot)  

plt.show()
```
